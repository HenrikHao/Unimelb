```python
import matplotlib.pyplot as plt
import pandas as pd

# Path to the file containing the depth information
base_file_path = 'minimax_base_output.txt'
advanced_file_path = 'minimax_advanced_output.txt'

# Read the depth values from the file
base_depths = []
with open(base_file_path, 'r') as file:
    for line in file:
        if 'depth' in line:
            # Extract the number after "depth"
            depth_value = int(line.strip().split(' ')[1])
            base_depths.append(depth_value)

advanced_depths = []
with open(advanced_file_path, 'r') as file:
    for line in file:
        if 'depth' in line:
            # Extract the number after "depth"
            depth_value = int(line.strip().split(' ')[1])
            advanced_depths.append(depth_value)

df_base = pd.DataFrame(base_depths, columns=['Basic Evaluation'])
df_advanced = pd.DataFrame(advanced_depths, columns=['Advanced Evaluation'])

# Concatenate the dataframes along the columns
combined_df = pd.concat([df_base, df_advanced], axis=1)

# Create a boxplot
plt.figure(figsize=(10, 6))
combined_df.boxplot()
plt.title('Comparison of Depth Information')
plt.ylabel('Depth Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Path to the file containing the depth information
mcts_file_path = 'mcts_time_output.txt'
minimax_file_path = 'minimax_time_output.txt'
greedy_file_path = 'yixiang_greedy_time_output.txt'

# Read the depth values from the file
mcts_time = []
with open(mcts_file_path, 'r') as file:
    for line in file:
        if 'Time: ' in line:
            # Extract the number after "depth"
            time_value = float(line.strip().split('  ')[1])
            mcts_time.append(time_value)

minimax_time = []
with open(minimax_file_path, 'r') as file:
    for line in file:
        if 'Time: ' in line:
            # Extract the number after "depth"
            time_value = float(line.strip().split('  ')[1])
            minimax_time.append(time_value)

greedy_time = []
with open(greedy_file_path, 'r') as file:
    for line in file:
        if 'Time: ' in line:
            # Extract the number after "depth"
            time_value = float(line.strip().split('  ')[1])
            greedy_time.append(time_value)

df_mcts = pd.DataFrame(mcts_time, columns=['MCTS Time'])
df_minimax = pd.DataFrame(minimax_time, columns=['MiniMax Time'])
df_greedy = pd.DataFrame(greedy_time, columns=['Greedy Time'])

# Concatenate the dataframes along the columns
combined_df = pd.concat([df_mcts, df_minimax, df_greedy], axis=1)

# Create a boxplot
plt.figure(figsize=(10, 6))
combined_df.boxplot()
plt.title('Comparison of Time Information')
plt.ylabel('Time Values')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Distribution of the dataset

```python
full_train_data = pd.read_csv("../train.csv")
```

```python
full_train_data.columns
```

```python
import matplotlib.pyplot as plt

def plot_distribution(df, field):
    distribution = df[field].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    plt.bar(distribution.index, distribution.values)
    
    # Setting font sizes
    plt.xlabel(field, fontsize=18)  # Increase font size for x-axis label
    plt.ylabel('Frequency', fontsize=18)  # Increase font size for y-axis label
    plt.title('Distribution of ' + field, fontsize=18)  # Increase font size for title
    
    # Set tick parameters
    plt.xticks(fontsize=18)  # Increase font size for x-axis ticks
    plt.yticks(fontsize=18)  # Increase font size for y-axis ticks
    
    plt.show()
```

```python
height_distribution = full_train_data['stable_height'].value_counts().sort_index()
```

```python
# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(height_distribution.index, height_distribution.values)
plt.xlabel('Stable Height')
plt.ylabel('Frequency')
plt.title('Distribution of Stable Height')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
image_dir = '../train/'
full_train_data['filename'] = full_train_data['id'].apply(lambda x: f"{image_dir}{x}.jpg")
train_df, val_df = train_test_split(full_train_data, test_size=0.2, random_state=42)
```

```python
plot_distribution(train_df, 'stable_height')
```

```python
plot_distribution(val_df, 'stable_height')
```
